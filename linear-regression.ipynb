What is Linear Regression?
<br>
It is a stastical method of modeling one or more dependent variables to one dependend variable.
Confusing?
okay, understand it in simple method....Let's say, you have 15 Apples in a basket, you and your friend are only two in the room to pic all thes Apples. Now you picked the 7 apples....how many will be picked by your friend (given he can pick as many as possible)? he will have 15-7=8 Apples.

Here you were independent to pick any number of apples, but your friend was depended on you (about how many you leave for him). This relationship can be modeled as an equation of line, i.e

Y = mX + c  or Y = wX + b (where W is weights and b is bias.)

Multiple linear regression ?
<br>
Here is the catch here, we have multiple indiependent variable to incoorporate, in this case equation will be:

Y = w1X1 + w2X2 + w3X3 + w4X4 + ....................+wnXn + b

How does it work ?
<br>
The algorithm tries to find the best-fitting line by minimizing the sum of squared errors (the vertical distance between the actual points and the predicted line). This method is called Ordinary Least Squares (OLS).

lets see the maths here...

if your equation is Y = wX + b for any give X you have Y-predicted and Y-Actual.
Now to calculate error, we need to see the diffrence between actual and predicted value i.e <br>

Error = Y-predicted - Y-Actual

to get OLS or Ordinary Least Squares we will do Sum(Error1^2 + Error2^2 + Error3^2 .....+ ErrorN^2)



Mean - Square - Error
<br>
MSE= OLS/N

```python
#Importing packages
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define Data
X = np.array([150, 160, 170, 180, 190]).reshape(-1, 1)  # independent variable
y = np.array([50, 60, 70, 80, 90])  # dependent variable

# let's plot to see the relationship between the datapoints

plt.scatter(X, y, color='blue')  
plt.plot(X, y, color='blue') 
```

```python
#Let's train a linear regression model on this
model = LinearRegression()
model.fit(X=X,y=y)
```

```python
ypred = model.predict(X=X)
```

```python
plt.scatter(X, y, color='blue')  
plt.plot(X, y, color='blue')
plt.scatter(X, ypred, color='red')  
plt.plot(X, ypred, color='red') 
```
